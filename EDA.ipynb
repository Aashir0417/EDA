```python
df = df.rename(columns = {
    'Draw' : 'Tie'
})

df =  df.drop("Goal Difference", axis=1)
df
```

```python
sns.countplot(x = "Win" , data=df)
```

```python
df["Goals Against"].value_counts()\
    .head(10) \
    .plot(kind="barh", title="GOALS Against world cup 2022")
```

```python

sns.heatmap(df.corr(), annot=True, cbar=True)
```

```python
plt.scatter(x=df["Win"] ,y=df["Loss"])
plt.show()
```

```python
ax= df['Goals Against'].plot(kind="hist", title= "Goals Against 2022 World cup")
ax.set_xlabel("Goals")
```

```python
ax= df['Goals Against'].plot(kind="pie", title= "Goals Against 2022 World cup")
ax.set_xlabel("Goals")
```

```python
df.plot(x= "Goals For" , y= "Goals Against" , kind="scatter")
plt.show()
```

```python
sns.scatterplot(x="Goals For" , y="Goals Against",hue = "Team", data=df)
```

```python
sns.pairplot(df, vars=["Goals For", "Goals Against", "Win", "Loss"])
plt.show()
```
